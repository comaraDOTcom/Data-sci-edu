## 7. Pandas Visualisations

This isn't that customisable compared to what matplotlib and Seaborn offer. However, it is very useful for quick data exploration 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

```python
import matplotlib as mpl
mpl.rcParams['patch.force_edgecolor'] = True
# This allows all the plots to take the Seaborn style as opposed to the plt standard.
```

```python
%matplotlib inline
```

```python
"""
The data files for this notebook are in a directory called "7_data"
"""
```

```python
cwd = os.getcwd()
```

```python
os.chdir(cwd+'/7_data')
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
# can plot dataframes and it uses matplotlib
df1['A'].hist(bins=30,color='green')
```

```python
df1['B'].plot(kind='hist')
```

```python
#Or equivalently
df1['B'].plot.hist()
```

```python
df2.plot.area(alpha=0.6)
```

```python
# Bar plot takes the index value as a category and plots all columns against it
df2.plot.bar(alpha=0.6)
```

```python
df2.plot.bar(stacked=True)
```

```python
df1 = df1.reset_index()
# Had to convert index from string to datetime
df1['index'] = pd.to_datetime(df1['index'])
df1.plot.line(x='index',y='B',figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=40,cmap='rainbow',figsize=(9,6))
```

```python
df2.plot.kde()
```
