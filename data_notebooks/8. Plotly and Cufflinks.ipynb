## Plotly and Cufflinks

Plotly is an interactive visualisation library and Cufflinks connects this to Pandas

Official documentation for Plotly is here (https://plot.ly/pandas/) 

Official documentation for Cufflinks is here (https://github.com/santosjorge/cufflinks)

```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
from plotly import __version__
print(__version__)
```

```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot, plot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
# DATA
df1 = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,56,76]})
```

```python
# interactive plot
df1.iplot()
```

```python
df1.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
# bar plot -- x-axis preferably a category
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
# to do a bar plot of df1, you want to do some aggregation on the dataframe first
df1.sum().iplot(kind='bar')
```

```python
 # to do a bar plot of df1, you want to do some aggregation on the dataframe first
df1.iplot(kind='box')
```

```python
# 3-dimensional dataframe
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y':[10,20,30,20,10], 'z':[5,4,3,2,1]})
```

```python
df3.iplot(kind='surface') #also takes colorscale argument (similar to cmap)
```

```python
df1['A'].iplot(kind='hist',bins=25)
```

```python
# Spread is good for financial data
df1[['A','B']].iplot(kind='spread')
```

```python
# Bubble plot is a scatter plot where the size of the bubble is determined by another column
df1.iplot(kind='bubble',x='A',y='B',size='C')

# these are common for World GDP vs happiness vs per capita 
```
