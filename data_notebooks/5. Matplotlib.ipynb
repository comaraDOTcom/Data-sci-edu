```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

## (A) Functional Method

```python
# FUNCTIONAL METHOD
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel( 'Y = X**2')
plt.title('F(X) = X**2')
plt.show()
```

```python
#plt.subplot(no.rows, no.columns, plot referring to)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

## (B) Object Oriented Method

```python
# OBJECT ORIENTED METHOD
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'*y-')
axes2.plot(y,x,'r*-')
axes1.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

## (C) Subplots

```python
fig, axes = plt.subplots(1,2)
# axes is a list of axes...ie. we can iterate and index it
for current_ax in axes:
    current_ax.plot(x,y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('1st plot') #call methods off axes subplot objects
axes[1].plot(x,y**2)
```

## (D) Figure Size and DPI (Dots per Inch)

```python
fig = plt.figure(figsize = (8,4)) # height by width in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1,figsize = (8,4)) # height by width in inches

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('Matplotlib_fig.png', dpi=600)
```

## (E) Legend

```python
fig = plt.figure(figsize = (8,4),dpi=300) # height by width in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='Y=X')
ax.plot(x,x*(1/2),label='Y=SQRT(X)')
ax.legend(loc)
```

## (F) Customisations

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color='purple')

# can use RGB hexadecimal codes too
ax.plot(x,y-2,color='#F0A321',lw=3, alpha=0.5,marker='o',markersize=10,markerfacecolor='green', markeredgewidth=2,
       markeredgecolor = 'black')
#alpha is the transparency of the line
ax.plot(x,y+2,color='red',lw=3,alpha=0.2, ls = 'steps')
# LS = Linestyle

# Setting limit on the ranges of the axes
ax.set_xlim([0,2])
ax.set_ylim([-5,5])
```
